## 1. Vectores y matrices
Un vector es una lista de números que representa información de un objeto.
Una matriz organiza datos en filas y columnas, y su producto permite calcular relaciones entre objetos y características.

```python
import numpy as np

#biblioteca de Python diseñada para trabajar con datos numéricos de forma eficiente.
#arrays, operaciones matemáticas, estadísticas, slice compatible con otras librerías como matplotlib

# Vector ejemplo: edades de usuarios
edades = np.array([25, 30, 22, 35])
print('Vector de edades:', edades)

# Matriz ejemplo: usuarios vs productos (puntuaciones)
A = np.array([[1,0,1],[0,1,1],[1,1,0]])  # usuarios vs características
B = np.array([[5,0,2],[0,4,1],[3,5,0]])  # características vs productos
C = np.dot(A,B)  # producto matricial
D= A + B
print('Producto matricial (predicción de puntuaciones):\n', C)
print("\n Suma de arrays:\n", D)
```

## 2. Media, mediana, moda y varianza
Estas son medidas estadísticas que resumen un conjunto de datos:
- Media: promedio
- Mediana: valor central
- Moda: valor más frecuente
- Varianza: dispersión respecto a la media

```python
import numpy as np
from scipy import stats

# Datos ejemplo: ventas de productos en una tienda
ventas = np.array([10, 15, 10, 20, 25, 10, 30])

# Media
media = round(np.mean(ventas), 2)

# Mediana
mediana = round(np.median(ventas), 2)

# Moda (solo el valor más frecuente)
moda = round(stats.mode(ventas, keepdims=True).mode[0], 2)

# Varianza
varianza = round(np.var(ventas), 2)

# Desviacion Standart
desviacion = round(np.std(ventas), 2)

print('Media:', media)
print('Mediana:', mediana)
print('Moda:', moda)
print('Varianza:', varianza)
print('DesVarianza:', desviacion)
```

## 3. Distribución de Poisson
Modela eventos discretos que ocurren con una tasa media en un intervalo (ej: número de clientes por minuto).

```python
import matplotlib.pyplot as plt
from scipy.stats import poisson

#Matplotlib es una biblioteca de Python para crear gráficos y visualizaciones.
#SciPy es una biblioteca de Python para computación científica y técnica, y se basa en NumPy. 
#Mientras que NumPy te da los arrays y operaciones básicas, SciPy añade un montón de herramientas avanzadas para matemáticas, ciencia e ingeniería.

lam = 4  # clientes promedio por minuto
x = np.arange(0,10)
y = poisson.pmf(x, lam)    #eventos que ocurren de forma aleatoria en un intervalo fijo de tiempo o espacio.
plt.bar(x,y, color= '#FF9999')
plt.title('Distribución Poisson (λ=4)')
plt.xlabel('Número de clientes')
plt.ylabel('Probabilidad')
plt.show()
```

## 4. Distribución Normal
Modela datos continuos agrupados alrededor de una media. Ej: altura o peso de personas.

```python
from scipy.stats import norm
mu, sigma = 170, 10  # media de altura y desviación estándar (cm)
 # mu = media de la distribución (promedio de la altura, en este caso).

#sigma = desviación estándar (qué tan dispersos están los datos respecto a la media).
x_norm = np.linspace(140, 200, 100)     #array de 100 puntos entres 140 y 200
y_norm = norm.pdf(x_norm, mu, sigma)   #calcula la función de densidad de probabilidad (PDF) de la distribución normal.
plt.plot(x_norm, y_norm)
plt.title('Distribución Normal (altura)')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad de probabilidad')
plt.show()
```

## 5. Derivada y regresión lineal
La derivada indica la tasa de cambio de una función.
La regresión lineal permite hacer predicciones, determinar los max y min haciendo la primera derivada 

```python
import numpy as np
import matplotlib.pyplot as plt

# Función cuadrática
x = np.linspace(-10, 10, 100)
y = x**2 + 2*x + 1

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Ejemplo de función cuadrática")
plt.grid(True)
plt.show()
```

Queremos predecir la nota de un alumno según las horas de estudio.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Datos: horas de estudio vs nota
horas = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
notas = np.array([2, 4, 5, 4, 5])

# Crear y entrenar modelo
modelo = LinearRegression()
modelo.fit(horas, notas)

# Parámetros de la recta
pendiente = round(modelo.coef_[0],2)
interseccion = round(modelo.intercept_,2)

# Predicción
nota_6h = round(modelo.predict(np.array([[6]]))[0],2)

print('Pendiente (m):', pendiente)
print('Intersección (b):', interseccion)
print('Predicción de nota para 6h de estudio:', nota_6h)
```

## 5. Derivada

Derivada de una función para cálcular la pendiente de la recta tangente a la curva.
Situación real: velocidad de un coche.

Supongamos que la posición del coche viene dada por:

s(t)= t^2 (s en metros, t en segundos)

La derivada

s'(t)= 2t  (t) nos dice la velocidad en cada instante.

```python
import numpy as np
import matplotlib.pyplot as plt

# Posición del coche
def s(t):
    return t**2

# Instantes de tiempo
t = np.linspace(0, 5, 100)
posicion = s(t)

# Velocidad aproximada en t = 3 s
t0 = 3
dt = 0.01
velocidad = (s(t0 + dt) - s(t0)) / dt

print(f"Velocidad aproximada en t={t0} s: {velocidad:.2f} m/s")

# Dibujar posición y velocidad
plt.plot(t, posicion, label="Posición s(t) = t^2")
plt.scatter(t0, s(t0), color='red', label=f"t={t0} s")
plt.arrow(t0, s(t0), 0.5, velocidad*0.5, color='green', width=0.05, label="Velocidad aprox")
plt.xlabel("Tiempo (s)")
plt.ylabel("Posición (m)")
plt.title("Derivada como velocidad")
plt.grid(True)
plt.legend()
plt.show()
```
