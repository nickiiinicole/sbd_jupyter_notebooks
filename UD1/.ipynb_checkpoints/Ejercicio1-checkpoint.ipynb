#### 1. Una tienda online quiere predecir la puntuación que cada usuario daría a un producto.

        Matriz de usuarios    U → cuánto les gusta cada característica de los productos.
        
        Matriz de productos   P → puntuación de cada característica de cada producto.
        
        #### Matriz de usuarios (filas=usuarios, columnas=características)
        U = np.array([
            [5, 3, 1],  # Usuario 1
            [2, 4, 5],  # Usuario 2
            [3, 5, 2]   # Usuario 3
        ])
        
        ##### Matriz de productos (filas=características, columnas=productos)
        P = np.array([
            [1, 0, 2],  # Característica 1
            [0, 1, 1],  # Característica 2
            [1, 2, 0]   # Característica 3
        ]) 
        
        Calcula El producto de matrices U×P que nos da una predicción de puntuaciones. Intenta explicar el resultado

- - - - - - - - -

```python
#primero importo las librerias necesarias : 
import numpy as np
#declaro las matrices 
#matriz de usuarios 
U = np.array([
        [5, 3, 1],  # Usuario 1
        [2, 4, 5],  # Usuario 2
        [3, 5, 2]   # Usuario 3
    ])
#matriz de productos 
P = np.array([
        [1, 0, 2],  # Característica 1
        [0, 1, 1],  # Característica 2
        [1, 2, 0]   # Característica 3
    ]) 

result = np.dot(U,P) # producto matricial 
print(f"El valor de producto matricial es \n {result}")
```

#### 2. Una tienda online registra las ventas diarias de un producto durante 30 días (1 mes) expuestos en el vector siguiente. Calcula media, mediana, moda y varianza.

```python
import numpy as np
from scipy import stats

ventas = [1057 , 662 ,1999 ,1690 ,1367,  931 ,1628, 1254, 
1608, 1942,  746,  120 , 940,  266, 1397,  487,  700 , 415 , 113 , 341,  876, 1469 , 664,  997 ,1463  
,191 ,1490 ,1055,  1578, 1551]

media = round(np.mean(ventas), 2) # media y redondea 2 decimales
mediana = round(np.median(ventas), 2) 
''' stats.mode() -> calcula el valor que mas se repite
        keepdims= True -> mantiene = forma del input
        .mode[0] -> devuelve un obj que tiene varias propiedas en este toamamos el primer valor en caso de que haya mas de una moda
'''
moda = round(stats.mode(ventas, keepdims=True).mode[0], 2)
varianza = round(np.var(ventas), 2)
print(f"La media es de : {media} \nLa mediana es de: {mediana} \nLa moda es de: {moda} \nLa varianza es de: {varianza}")
```

### 3. Calcula la distribución de Poisson y Normal del apartado anterior.
Calcula el número de días que superan los 1000 euros de ventas la probabilidad de haya más de 5 días con más de 1000 euros por día

Nota.- Para calcular el número de días con más 1000 euros ventas ====> dias_1000 = (ventas > 1000).su()

Muestra las gráficas correspondientes y explícalas

```python
#importar librerías
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson 

ventas = np.array([1057, 662, 1999, 1690, 1367, 931, 1628, 1254, 
                   1608, 1942, 746, 120, 940, 266, 1397, 487, 700, 
                   415, 113, 341, 876, 1469, 664, 997, 1463, 
                   191, 1490, 1055, 1578, 1551])

dias_1000 = (ventas > 1000).sum() # los dias que superan esa ventas 
print(f"Los dias  con ventas > 1000 : {dias_1000}")
# λ = número promedio de eventos (días con más de 1000 €)
# Como tenemos un solo mes, λ = dias_1000 directamente
lam = dias_1000
# Rango de posibles valores 30 dias
x = np.arange(0, 30)
y = poisson.pmf(x, lam)
plt.bar(x, y, color='#FF9999')
plt.title(f'Distribución de Poisson (λ={lam})')
plt.xlabel('Número de días con ventas > 1000 €')
plt.ylabel('Probabilidad')
plt.show()
```
